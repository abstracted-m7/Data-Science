```python
dummy=pd.get_dummies(read_file['f35'])
dummy.head()
```

```python
read_file2=pd.concat((read_file,dummy),axis=1)
read_file2.head(5)
```

```python
read_file2=read_file2.drop(['g'],axis=1)
read_file2.head(5)
```

```python
read_file2=read_file2.drop(['f35'],axis=1)
read_file2.head(5)
```

# Feature Engineering

```python
read_file2.isna().sum()
```

## Feature Selection

```python
read_file2.columns
```

```python
X=read_file2.drop('b',axis=1)
y=read_file2['b']
y.head()

```

```python
y.unique()
```

```python
import seaborn as sns
corr=X.corr()
corr.shape
```

```python
y.head()
```

```python
#Plotting heatmap to find correalation among features
plt.figure(figsize=(20,15))
sns.heatmap(corr,vmin=0,vmax=1,annot=True,fmt='.1f',cbar=True,square=True,cmap='coolwarm')
plt.show()
```

```python
def correlation(dataset,threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>threshold:
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr            
```

```python
corr_features=correlation(X,.5)
print(corr_features)
```

```python
X_new=X.drop(corr_features,axis=1)
X_new
```

```python
y.shape
```

```python
y.unique()
```

## Split the training and test data

```python
from sklearn.model_selection import train_test_split
X_new_train,X_new_test,y_train,y_test=train_test_split(X_new,y,test_size=.25,random_state=1)
```

```python
y_train.shape
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
naive=GaussianNB()
naive.fit(X_new_train,y_train)
```

```python
predicted_naive=naive.predict(X_new_test)
predicted_naive
```

```python
import sklearn.metrics as metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
cm_naive=metrics.confusion_matrix(y_test,predicted_naive)
cmd=ConfusionMatrixDisplay(cm_naive,display_labels=['1','0'])
cmd.plot()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predicted_naive))
```

```python
from sklearn.metrics import roc_curve
naive_probs=naive.predict_proba(X_new_test)
naive_probs=naive_probs[:,1]
naive_fpr,naive_tpr,_=roc_curve(y_test,naive_probs)
plt.plot(naive_fpr,naive_tpr,marker='.',label='Naive bayes')
plt.xlabel('False positive only')
plt.ylabel('True positive only')
plt.legend()
plt.show()
```

```python
accuracy_naive=metrics.accuracy_score(y_test,predicted_naive)
accuracy_naive
```
