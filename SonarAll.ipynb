# Feature Engineering

```python
read__file['f61']
```

```python
dummy=pd.get_dummies(read__file['f61'])
dummy.head()
```

```python
read_file2=pd.concat((read__file,dummy),axis=1)
read_file2.head()
```

```python
read_file2=read_file2.drop(['f61'],axis=1)
```

```python
read_file2=read_file2.drop(['M'],axis=1)
read_file2
```

## Feature Selection

```python
X=read_file2.drop('R',axis=1)
y=read_file2['R'].values
#y = dataset.iloc[:, 2].values
```

```python
import seaborn as sns
corr=X.corr()
corr.shape
```

```python
#Plotting Heatmap to find corelation among features
plt.figure(figsize=(30,20))
sns.heatmap(corr,vmin=0,vmax=1,annot=True,fmt='.1f',cbar=True,square=True,cmap='coolwarm')
plt.show()
```

```python
def correlation(dataset,threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>threshold:
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features=correlation(X,.6)
print(corr_features)
```

```python
X_new=X.drop(corr_features,axis=1)
X_new
```

```python
y.shape
```

## Splitting training and test data

```python
from sklearn.model_selection import train_test_split
#y=y.reshape(-1,1)
X_new_train,X_new_test,y_train,y_test=train_test_split(X_new,y,test_size=.25,random_state=1)

y_train.shape

```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
naive=GaussianNB()
naive.fit(X_new_train,y_train)
```

```python
predicted_naive=naive.predict(X_new_test)
predicted_naive
```

```python
import sklearn.metrics as metrics
from sklearn.metrics import ConfusionMatrixDisplay
cm_naive=metrics.confusion_matrix(y_test,predicted_naive)
cmd=ConfusionMatrixDisplay(cm_naive,display_labels=['1','0'])
cmd.plot()

```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predicted_naive))
```

```python
accuracy_naive=metrics.accuracy_score(y_test,predicted_naive)
print(accuracy_naive)
```

```python
naive_probs=naive.predict_proba(X_new_test)
naive_probs=naive_probs[:,1]
from sklearn.metrics import roc_curve
naive_fpr,naive_tpr,_=roc_curve(y_test,naive_probs)
plt.plot(naive_fpr,naive_tpr,marker='.',label='Naive Bayes')
plt.xlabel('False Positive only')
plt.ylabel('True Positive only')
plt.legend()
plt.show()
```
